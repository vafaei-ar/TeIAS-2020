```python
%matplotlib inline

import numpy as np
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x=tips["total_bill"])
```

```python
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
# Can do entire dataframe with orient='h'
sns.boxplot(data=tips,palette='rainbow',orient='h')
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm")
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',palette='Set1')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',split=1,palette='Set1')
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=0)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1')
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1',dodge=True)
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.violinplot(x="tip", y="day", data=tips,palette='rainbow')
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=3)
```

```python
sns.set()
```

```python
sns.distplot(tips['total_bill'])
# Safe to ignore warnings
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
import matplotlib.pyplot as plt

# Map to upper,lower, and diagonal
g = sns.PairGrid(tips)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
sns.heatmap(tips.corr(),cmap='coolwarm',annot=True)
```
